# Simular cero y pendiente  para acondicionamiento de señales

- Se generará una función para emular el proceso de cero y pendiente en el acondicionamiento de señales.

- Las funciones principales para generar vectores equiespaciados son [linspace](https://numpy.org/doc/stable/reference/generated/numpy.linspace.html), [range](https://www.w3schools.com/python/ref_func_range.asp) y [arange](https://numpy.org/doc/stable/reference/generated/numpy.arange.html)

```
# crear vector tiempo por rango y por número de muestras
K = 200 # número de muestras
ti = -15 # tiempo incial referencia [s]
tf = 10 # tieimpo final  [s]
# vector de tiempo equiespacioado
tv = np.linspace(start=ti,stop=tf,num=K)
print(tv)
print(tv.shape)
# tiempo muestreo se define según el rango de trabajo en segundos y la cantidad de muestras
Ts = (tf - ti)/K
print(Ts)
```

$x[t_k] = a t_k^2 + b$





```
#parámetros parabola
a = 4
b = -3
xv = a*(tv**2) + b # creamos parabola en tiempo discreto-> **  pontencia
print(xv)
print(xv.shape)
```

## Cálculo de pendiente y del offset (cero)

$$ y[t_k] \in [y_{min},y_{max}]$$

$$y[t_k] = m x[t_k] + c$$

$$m = \frac{y_{max}-y_{min}}{x_{max}-x_{min}}$$

$$c = y[t_k] - m x[t_k]$$

- Se crea un código tipo función bajo el ambiente def, para generar el cero y pendiente según los valores $y_{max}$ y $y_{min}$ definidos por el usuario.

```
y_min = 0 # valor min después de cero y pendiente
y_max = 5 # valor max después de cero y pendiente
#crear funcion en python
def my_ceropen(x,ymin = -5, ymax = 5): #en general se pueden definir valores por defecto
  '''
  Codigo base para simular proceso de cero y pendiente
  Se ingresa arreglo de numpy y los valores min  y max despues de cero y pendiente
  '''
  xmax = max(x) # x.max()
  xmin = min(x) # x.min()
  m = (ymax - ymin)/(xmax - xmin)
  c = ymin - m*xmin
  yv = m*x + c
  return yv
```

```
yv = my_ceropen(xv,ymin=-15,ymax=20) #se llama la función definida en la celda anterior
```

```
print(yv)
print(yv.shape)
```

Podemos crear una función de cuantización y evaluarla punto a punto a partir de cada muestra después de cero y spam discretizada

```
#función cuantización punto a punto

def my_cuantizador(yn, vq) : #yn punto a #cuantizar, vq vector de estados

  Ne = vq.shape[0] #tamaño vector de estados
  dn = cdist(yn.reshape(-1,1),vq.reshape(-1,1))#distancia yn a vector estados, reshape(-1,1) asegura vectores columna para poder utilizar cdist
  ind = np.argmin(dn) #posición distancia min
  return vq[ind]
```

```
#recorrer vector cero y spam discretizado para cuantizar

for i in range(ycs.shape[0]):
   if i==0: #primera muestra
     yq = my_cuantizador(ycs[i],ve)
   else:
     yq = np.append(yq,my_cuantizador(ycs[i],ve)) #apilar datos cuantizando si i>0
```

```
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la señal
a = 4
b = -3
F0 = 60  # Frecuencia de la señal original en Hz
Fs = 2 * F0  # Frecuencia de muestreo mínima según criterio Fs ≥ 2F0
Ti, Tf = 0, 1  # Intervalo de tiempo

# Definición del tiempo continuo y la señal original
t_continuo = np.linspace(Ti, Tf, 1000)  # Señal continua con alta resolución
x_continuo = (a * t_continuo) ** 2 + b

# Definición del tiempo discreto y la señal muestreada
t_discreto = np.arange(Ti, Tf, 1 / Fs)  # Puntos de muestreo
x_discreto = (a * t_discreto) ** 2 + b

# Graficación
plt.figure(figsize=(10, 5))

# Señal continua
plt.plot(t_continuo, x_continuo, label="Señal continua x(t)", linestyle="-")

# Señal discretizada
plt.stem(t_discreto, x_discreto, linefmt="r-", markerfmt="ro", basefmt="k-", label="Señal discretizada x[n]")

# Configuración del gráfico
plt.xlabel("Tiempo (s)")
plt.ylabel("Amplitud")
plt.title("Discretización de la señal con Fs ≥ 2F0")
plt.legend()
plt.grid()
plt.show()
```

Realice una función mediante el ambiente def que permita realizar los procesos de discretización y cuantización (usando cdist y argmin) en cascada para un conversor análogo-digital de 3 bits con entrada análoga de 0 a 10 [v]. Presente las gráficas de la señal original, señal con cero y pendiente, señal discretizada, y señal digital.**

```
from scipy.spatial.distance import cdist

def adc_simulation(frequency=50, sampling_rate=200, duration=0.1):
    """
    Simula un conversor ADC de 3 bits con discretización y cuantización en cascada.

    Parámetros:
     frequency: Frecuencia de la señal analógica (Hz)
     sampling_rate: Frecuencia de muestreo (Hz), debe ser mayor al doble de la frecuencia
    duration: Duración de la señal en segundos
    """

    # Generar la señal original (senoidal en el rango de 0 a 10 V)
    t = np.linspace(0, duration, int(sampling_rate * duration))
    signal_analog = 5 * (1 + np.sin(2 * np.pi * frequency * t))  # Normalizada en [0, 10]

    # Aplicar "cero y pendiente" (normalización)
    signal_normalized = signal_analog / 10

    # Discretización (muestreo)
    t_sampled = np.arange(0, duration, 1 / sampling_rate)
    sampled_signal = 5 * (1 + np.sin(2 * np.pi * frequency * t_sampled))

    # Cuantización con ADC de 3 bits (8 niveles)
    levels = np.linspace(0, 10, 8).reshape(-1, 1)  # Niveles de cuantización
    sampled_signal_reshaped = sampled_signal.reshape(-1, 1)
    closest_indices = np.argmin(cdist(sampled_signal_reshaped, levels, metric='euclidean'), axis=1)
    quantized_signal = levels[closest_indices].flatten()  # Señal cuantizada

    # Conversión a señal digital (valores en binario)
    digital_signal = [format(idx, '03b') for idx in closest_indices]

    # Graficación
    plt.figure(figsize=(12, 8))

    # Gráfico 1: Señal original
    plt.subplot(2, 2, 1)
    plt.plot(t, signal_analog, label="Señal original", linestyle="-")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Voltaje (V)")
    plt.title("Señal original")
    plt.grid()

    # Gráfico 2: Señal con cero y pendiente (normalizada)
    plt.subplot(2, 2, 2)
    plt.plot(t, signal_normalized, label="Señal normalizada", linestyle="-", color='g')
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Voltaje Normalizado")
    plt.title("Señal con cero y pendiente")
    plt.grid()

    # Gráfico 3: Señal discretizada
    plt.subplot(2, 2, 3)
    plt.stem(t_sampled, sampled_signal, linefmt="r-", markerfmt="ro", basefmt="k-", label="Señal discretizada")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Voltaje (V)")
    plt.title("Señal discretizada")
    plt.grid()

    # Gráfico 4: Señal digital
    plt.subplot(2, 2, 4)
    plt.stem(t_sampled, quantized_signal, linefmt="b-", markerfmt="bo", basefmt="k-", label="Señal cuantizada")
    plt.xlabel("Tiempo (s)")
    plt.ylabel("Voltaje (V)")
    plt.title("Señal digital (cuantizada)")
    plt.grid()

    plt.tight_layout()
    plt.show()

# Ejecutar la simulación del ADC de 3 bits
adc_simulation()
```
